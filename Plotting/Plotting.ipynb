```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

data = pd.read_csv("C:/Users/Maruata/Documents/Anaconda/AEDADMSE/performance.csv")
```

```python
history_dict = data

insdn_dnn_accuracy = history_dict['insdn_dnn_accuracy']
insdn_dnn_val_accuracy = history_dict['insdn_dnn_val_accuracy']
insdn_dnn_loss = history_dict['insdn_dnn_loss']
insdn_dnn_val_loss = history_dict['insdn_dnn_val_loss']

insdn_dcnn_accuracy = history_dict['insdn_dcnn_accuracy']
insdn_dcnn_val_accuracy = history_dict['insdn_dcnn_val_accuracy']
insdn_dcnn_loss = history_dict['insdn_dcnn_loss']
insdn_dcnn_val_loss = history_dict['insdn_dcnn_val_loss']

cicids2017_dnn_accuracy = history_dict['cicids2017_dnn_accuracy']
cicids2017_dnn_val_accuracy = history_dict['cicids2017_dnn_val_accuracy']
cicids2017_dnn_loss = history_dict['cicids2017_dnn_loss']
cicids2017_dnn_val_loss = history_dict['cicids2017_dnn_val_loss']

cicids2017_dcnn_accuracy = history_dict['cicids2017_dcnn_accuracy']
cicids2017_dcnn_val_accuracy = history_dict['cicids2017_dcnn_val_accuracy']
cicids2017_dcnn_loss = history_dict['cicids2017_dcnn_loss']
cicids2017_dcnn_val_loss = history_dict['cicids2017_dcnn_val_loss']

cicddos2019_dnn_accuracy = history_dict['cicddos2019_dnn_accuracy']
cicddos2019_dnn_val_accuracy = history_dict['cicddos2019_dnn_val_accuracy']
cicddos2019_dnn_loss = history_dict['cicddos2019_dnn_loss']
cicddos2019_dnn_val_loss = history_dict['cicddos2019_dnn_val_loss']

cicddos2019_dcnn_accuracy = history_dict['cicddos2019_dcnn_accuracy']
cicddos2019_dcnn_val_accuracy = history_dict['cicddos2019_dcnn_val_accuracy']
cicddos2019_dcnn_loss = history_dict['cicddos2019_dcnn_loss']
cicddos2019_dcnn_val_loss = history_dict['cicddos2019_dcnn_val_loss']

start_by_epoch = 1
epochs = range(start_by_epoch, len(insdn_dcnn_accuracy) + 1)

sn.set(font_scale=2.35)

fig, ax = plt.subplots(figsize=(15,15))
ax.set_facecolor("white")
plt.plot(epochs, insdn_dcnn_accuracy[start_by_epoch-1:], '-', linewidth=3)
plt.plot(epochs, insdn_dcnn_val_accuracy[start_by_epoch-1:], '*-', linewidth=3)
plt.plot(epochs, cicids2017_dcnn_accuracy[start_by_epoch-1:], '--', linewidth=3)
plt.plot(epochs, cicids2017_dcnn_val_accuracy[start_by_epoch-1:], 'o--', linewidth=3)
plt.plot(epochs, cicddos2019_dcnn_accuracy[start_by_epoch-1:], '--', linewidth=3)
plt.plot(epochs, cicddos2019_dcnn_val_accuracy[start_by_epoch-1:], 'x--', linewidth=3)

plt.title('Comparison of Training and Testing Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.legend(['Accuracy: DCNN Trained with InSDN', 'Validation Accuracy: DCNN Trained with InSDN', 
           'Accuracy: DCNN Trained with CICIDS2017', 'Validation Accuracy: DCNN Trained with CICIDS2017',
           'Accuracy: DCNN Trained with CICDDoS2019', 'Validation Accuracy: DCNN Trained with CICDDoS2019'
           ], loc='lower right', facecolor='white', framealpha=1)

plt.grid(axis='y', alpha=.3, color='k', linestyle='-', linewidth=2)
plt.grid(axis='x', alpha=.2, color='k', linestyle='-', linewidth=2)
plt.ylim(0.9895,1.001)

plt.show()

fig.savefig('DCNN_accuracy.pdf') 

plt.clf()

fig, ax = plt.subplots(figsize=(15,15))
ax.set_facecolor("white")

plt.plot(epochs, insdn_dcnn_loss[start_by_epoch-1:], '-', linewidth=3)
plt.plot(epochs, insdn_dcnn_val_loss[start_by_epoch-1:], '*-', linewidth=3)
plt.plot(epochs, cicids2017_dcnn_loss[start_by_epoch-1:], '--', linewidth=3)
plt.plot(epochs, cicids2017_dcnn_val_loss[start_by_epoch-1:], 'o--', linewidth=3)
plt.plot(epochs, cicddos2019_dcnn_loss[start_by_epoch-1:], '--', linewidth=3)
plt.plot(epochs, cicddos2019_dcnn_val_loss[start_by_epoch-1:], 'x--', linewidth=3)

plt.title('Comparison of Training and validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.legend(['Loss: DCNN Trained with InSDN', 'Validation Loss: DCNN Trained with InSDN', 
            'Loss: DCNN Trained with CICIDS2017', 'Validation Loss: DCNN Trained with CICIDS2017',
            'Loss: DCNN Trained with CICDDoS2019', 'Validation Loss: DCNN Trained with CICDDoS2019'
           ], loc='upper right', facecolor='white', framealpha=1)

plt.grid(axis='y', alpha=.3, color='k', linestyle='-', linewidth=2)
plt.grid(axis='x', alpha=.2, color='k', linestyle='-', linewidth=2)
plt.ylim(0.001,0.0170)

plt.tight_layout()
plt.show()

fig.savefig('DCNN_Loss.pdf') 

plt.clf()
```
